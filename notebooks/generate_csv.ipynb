```python
import os
import pandas as pd

label_dir = 'data/train/labels'
image_dir = 'data/train/images'
output_csv = 'train_labels.csv'

data = []

for fname in os.listdir(label_dir):
    if not fname.endswith('.txt') or fname == 'classes.txt':
        continue

    label_path = os.path.join(label_dir, fname)
    with open(label_path, 'r') as f:
        line = f.readline().strip()
        if not line:
            continue  # skip empty

        # Use the first token as class name (not index)
        class_name = line.split()[0]

    # Match image filename
    image_name = os.path.splitext(fname)[0] + '.jpg'  # or .png if that’s what your images are
    if os.path.exists(os.path.join(image_dir, image_name)):
        data.append({'filename': image_name, 'class': class_name})
    else:
        print(f"Image not found: {image_name}")

# Save CSV
df = pd.DataFrame(data)
df.to_csv(output_csv, index=False)
print(f"Saved label mapping to {output_csv}")
```
